```python
#Importing Essential Libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
D = pd.read_csv("hiring.csv")
D
```

```python
#filling values of zero in experience column where NaN occurs.
D.experience = D.experience.fillna("zero")
D
```

```python
#converting the word number to numeric representation
D.experience = D.experience.apply(w2n.word_to_num)
D
```

```python
#putting mean value in test score column where NaN occurs by Ceil function
import math
median_test_score = math.ceil(D['test_score(out of 10)'].mean())
median_test_score
```

```python
D['test_score(out of 10)'] = D['test_score(out of 10)'].fillna(median_test_score)
D
```

```python
#fitting on three dependent variables
reg = linear_model.LinearRegression()
reg.fit(D[['experience','test_score(out of 10)','interview_score(out of 10)']],D['salary($)'])
```

```python
#1st exp, 2nd t_score, 3rd I_score
reg.predict([[2,9,6]])
```
